```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub


path = kagglehub.dataset_download("arsalanjamal002/student-sleep-patterns")
print("Path to dataset files:", path)
import os
files = os.listdir(path)
print("Files in folder:", files)
##Code provided from the Kaggle website^^

df = pd.read_csv(f"{path}/{files[0]}")
df.head()
#Table below
```

Interesting Statistics:

```python
# 1. Percentage of Students that sleep > 7 Hours
percentile_sleep_gt_7 = (df['Sleep_Duration'] > 7).mean() * 100

# 2. Amount of Students that Study Hours > Sleep Duration
study_gt_sleep_count = (df['Study_Hours'] > df['Sleep_Duration']).sum()

# 3. Mean and Median Screen Time
mean_screen_time = df['Screen_Time'].mean()
median_screen_time = df['Screen_Time'].median()

# 4. Caffeine Usage (Mean and Median)
mean_caffeine = df['Caffeine_Intake'].mean()
median_caffeine = df['Caffeine_Intake'].median()

print("4 Interesting Stats:")
print(f"1. Students sleeping > 7 hours: {percentile_sleep_gt_7:.1f}%")
print(f"2. Students studying more than they sleep: {study_gt_sleep_count}")
print(f"3. Screen Time -> Mean: {mean_screen_time:.2f} hrs | Median: {median_screen_time:.2f} hrs")
print(f"4. Caffeine Usage -> Mean: {mean_caffeine:.2f} drinks | Median: {median_caffeine:.2f} drinks")
```

Boxplot visualizing sleep differences between students depending on their grade.

```python
plt.figure(figsize=(8, 6))

# Define the order
order = ['1st Year', '2nd Year', '3rd Year', '4th Year']

# ci=95 adds the error bars (95% confidence interval)
sns.barplot(data=df, x='University_Year', y='Sleep_Duration', order=order, palette='viridis', ci=95)

plt.title('Average Sleep Duration by University Year', fontsize=14)
plt.xlabel('University Year')
plt.ylabel('Average Sleep Duration (Hours)')

# Setting the y-axis limit helps emphasize small differences
plt.ylim(0, 10)

plt.tight_layout()
plt.show()
```
